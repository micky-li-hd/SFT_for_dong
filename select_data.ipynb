```python
import os

# 设置两个目录路径
tar_dir = "/home/v-haodongli/mnt/v-haodongli-container/cot_output_test"
log_dir = "/home/v-haodongli/mnt/v-haodongli-container/status"

# 用于存储有问题的 tar 文件
failed_log_files = []

# 获取所有 .log 文件
log_files = [f for f in os.listdir(log_dir) if f.endswith('.log')]

for log_file in log_files:
    base_name = os.path.splitext(log_file)[0]  # 去掉 .log，比如得到 "00001"
    log_file_path = os.path.join(log_dir, log_file)

    # 构造对应的 .tar 文件名
    tar_file = f"{base_name}.tar"
    tar_file_path = os.path.join(tar_dir, tar_file)

    # 检查 log 文件是否包含 "failed"
    try:
        with open(log_file_path, 'r', encoding='utf-8') as f:
            content = f.read()
            if "failed" in content.lower():
                failed_log_files.append(tar_file_path)
    except Exception as e:
        print(f"Error reading {log_file_path}: {e}")

# 输出结果
print("\nFailed log files:")
for f in failed_log_files:
    print(f)
```

```python
import os
import re

# 日志路径
LOG_DIR = "/mnt/v-haodongli/status"

# 正则提取所有 .log 文件名中的 shard_id（如 02901）
input_text = """
... 这里是你提供的日志内容 ...
"""

# 提取所有 shard_id（从日志中提取数字部分）
shard_ids = list(set(re.findall(r"(\d+)\.log", input_text)))
shard_ids.sort()

# 构建 base job 字段
base_job = {
    "sku": "1xG1-A100",
    "sla_tier": "Standard",
    "priority": "High",
    "submit_args": {
        "max_run_duration_seconds": 1209600,
        "env": {
            "_AZUREML_SINGULARITY_JOB_UAI": "/subscriptions/622ac093-cd5d-4146-af74-d19d1d59f314/resourcegroups/MSRA_IEG_Multimedia/providers/Microsoft.ManagedIdentity/userAssignedIdentities/msramig3dmanagedidentity"
        }
    },
    "command": ["SHARD_ID={shard} bash gen_data.sh"]
}

# 分组处理每个 shard_id
valid_jobs = []

for shard_id in shard_ids:
    log_path = os.path.join(LOG_DIR, f"{shard_id}.log")

    # 如果文件不存在 或者 存在但包含 failed
    if not os.path.exists(log_path):
        valid_jobs.append(shard_id)
    else:
        try:
            with open(log_path, "r") as f:
                content = f.read()
            if "failed" in content.lower():
                valid_jobs.append(shard_id)
        except Exception as e:
            print(f"[ERROR] 读取 {log_path} 失败: {e}")
            continue

# 输出结果到终端
print("jobs:")
for shard_id in valid_jobs:
    padded_shard = f"{int(shard_id):05d}"  # 补零成 00000 格式
    name = f"cot_gen_shard_{padded_shard}"
    command = [cmd.replace("{shard}", str(int(shard_id))) for cmd in base_job["command"]]

    print(f"- name: {name}")
    print(f"  sku: {base_job['sku']}")
    print(f"  sla_tier: {base_job['sla_tier']}")
    print(f"  priority: {base_job['priority']}")
    print(f"  submit_args:")
    print(f"    max_run_duration_seconds: {base_job['submit_args']['max_run_duration_seconds']}")
    print(f"    env:")
    print(f"      _AZUREML_SINGULARITY_JOB_UAI: {base_job['submit_args']['env']['_AZUREML_SINGULARITY_JOB_UAI']}")
    print(f"  command:")
    print(f"  - {command[0]}")
    print()
```
